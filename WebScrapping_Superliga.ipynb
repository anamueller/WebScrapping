#Functions

```python
#função que ao passar o conteúdo html adquirido pelo beautiful soup acha os nomes dos times
def times(content):
  for i in content.find_all("span", {"id": "Content_Main_LBL_HomeTeam"}): #acha o time da casa
    home_team=i.text
  for i in content.find_all("span", {"id": "Content_Main_LBL_GuestTeam"}): #acha o time visitante
    guest_team=i.text
  return home_team, guest_team
```

```python
#função que ao passar o conteúdo html adquirido pelo beautiful soup acha o resultado da partida
def results(content):
  for i in content.find_all("span", {"id": "Content_Main_LBL_WonSetHome"}): #time da casa sets vencidos
    home_team_sets = i.text
  for i in content.find_all("span", {"id": "Content_Main_LBL_WonSetGuest"}): #time visitante sets vencidos
    guest_team_sets = i.text
  return home_team_sets, guest_team_sets
```

```python
def sets(content):
  sets =[]
  for i in content.find_all("span", {"id": "Content_Main_ctl17_L_MatchAnalysis_SetBySet_Set1_Home"}): #div que contém as parciais de cada set
      sets.append(i.text) #set 1 home
  for i in content.find_all("span", {"id": "Content_Main_ctl17_L_MatchAnalysis_SetBySet_Set1_Guest"}): #div que contém as parciais de cada set
      sets.append(i.text) #set 1 guest
  for i in content.find_all("span", {"id": "Content_Main_ctl17_L_MatchAnalysis_SetBySet_Set2_Home"}): #div que contém as parciais de cada set
      sets.append(i.text) #set 2 home
  for i in content.find_all("span", {"id": "Content_Main_ctl17_L_MatchAnalysis_SetBySet_Set2_Guest"}): #div que contém as parciais de cada set
      sets.append(i.text) #set 2 guest
  for i in content.find_all("span", {"id": "Content_Main_ctl17_L_MatchAnalysis_SetBySet_Set3_Home"}): #div que contém as parciais de cada set
      sets.append(i.text) #set 3 home
  for i in content.find_all("span", {"id": "Content_Main_ctl17_L_MatchAnalysis_SetBySet_Set3_Guest"}): #div que contém as parciais de cada set
      sets.append(i.text) #set 3 guest
  for i in content.find_all("span", {"id": "Content_Main_ctl17_L_MatchAnalysis_SetBySet_Set4_Home"}): #div que contém as parciais de cada set
      sets.append(i.text) #set 4 home
  for i in content.find_all("span", {"id": "Content_Main_ctl17_L_MatchAnalysis_SetBySet_Set4_Guest"}): #div que contém as parciais de cada set
      sets.append(i.text) #set 4 guest
  for i in content.find_all("span", {"id": "Content_Main_ctl17_L_MatchAnalysis_SetBySet_Set5_Home"}): #div que contém as parciais de cada set
      sets.append(i.text) #set 5 home
  for i in content.find_all("span", {"id": "Content_Main_ctl17_L_MatchAnalysis_SetBySet_Set5_Guest"}): #div que contém as parciais de cada set
      sets.append(i.text) #set 5 guest
  return sets
```

```python
def jogadoras(content):
  jogadoras_time_1 = []
  jogadoras_time_2 = []
  jogadoras = []
  liberos = []
  #acha todas as jogadores da partida, reservas e titulares, de ambos os times
  for i in content.find_all("span", {"id": "PlayerName"}): #div com os nomes das jogadores
    if i.text in jogadoras:
      continue
    else:
      jogadoras.append(i.text)

  #acha as liberos
  for i in range(0, len(jogadoras)):
    nome_com_l = jogadoras[i]
    if nome_com_l[-4:] == ' (L)':
      nome_sem_l = nome_com_l[:len(nome_com_l)-4]
      liberos.append(nome_sem_l)

  #remove nomes repetidos das liberos
  for i in range(0, len(liberos)):
    jogadoras.remove(liberos[i])

  #separa as jogadores do time 1
  for i in range(0, len(jogadoras)):
    if jogadoras[i] == 'TOTAIS':
      break
    jogadoras_time_1.append(jogadoras[i]) 

  #separa as jogadores do time 2
  for i in range(len(jogadoras_time_1)+1, len(jogadoras)):
    jogadoras_time_2.append(jogadoras[i])

  return jogadoras_time_1, jogadoras_time_2
```

#Dataframe

```python
import pandas as pd
from tqdm import tqdm
```

```python
legend = ['ID partida', 'Home Team Name', 'Guest Team Name', 'Home Team Sets Won', 'Guest Team Sets Won', 'Home Team Set 1',
            'Home Team Set 2', 'Home Team Set 3', 'Home Team Set 4', 'Home Team Set 5', 'Guest Team Set 1', 'Guest Team Set 2',
           'Guest Team Set 3', 'Guest Team Set 4', 'Guest Team Set 5', 'Home Team Spikes', 'Guest Team Spikes', 'Home Team Aces',
           'Guest Team Aces', 'Home Team Blocks', 'Guest Team Blocks', 'Home Team Errors', 'Guest Team Errors', 'Home Team Opp Points',
           'Guest Team Opp Points', 'Home Team Outside Points', 'Guest Team Outside Points',  'Home Team MB Points', 'Guest Team MB Points',
          'Home Team Setters Points','Guest Team Setters Points']
```

```python
ID=[]
for i in range(0,len(links)):
  ID.append(i)
```

```python
dados_volei = pd.DataFrame(index=ID, columns=legend)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
for i in tqdm(range(112,len(links))):
  #getting the page source content before changing the page
  html_content = requests.get(links[i]).text
  estatistica_do_jogo = BeautifulSoup(html_content, "lxml")
  #getting the page source content after changing the page (clicking on the button analise de jogo)
  driver.get(links[i])
  driver.find_element(By.XPATH, '//span[text()="Analise do Jogo"]').click()
  html = driver.page_source
  analise_do_jogo = BeautifulSoup(html, "lxml")
  print(links[i])
  home_team, guest_team = times(estatistica_do_jogo)
  sets_home, sets_guest = results(estatistica_do_jogo)
  points_sets = sets(analise_do_jogo)
  home_spikes, guest_spikes = spikes(analise_do_jogo)
  home_aces, guest_aces = aces(analise_do_jogo)
  home_blocks, guest_blocks = bloqueios(analise_do_jogo)
  home_errors, guest_errors = errors(analise_do_jogo)
  home_opp, guest_opp = pontos_oposto(analise_do_jogo)
  home_out, guest_out = pontos_ponteiros(analise_do_jogo)
  home_mb, guest_mb = pontos_central(analise_do_jogo)
  home_sett, guest_sett = pontos_levantador(analise_do_jogo)

  teste = [id[i], home_team, guest_team, sets_home, sets_guest, points_sets[0], points_sets[2], points_sets[4], points_sets[6], points_sets[8],
          points_sets[1], points_sets[3], points_sets[5], points_sets[7], points_sets[9], home_spikes, guest_spikes, home_aces, guest_aces,
          home_blocks, guest_blocks, home_errors, guest_errors, home_opp, guest_opp, home_out, guest_out, home_mb, guest_mb, home_sett, 
          guest_sett]
        
  dados_volei.loc[i] = teste
```
